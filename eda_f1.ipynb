Task 2: Find the top 10 drivers by total career points. Plot it as a bar chart.

```python
# join results with drivers
resultsxdrivers_df = results_df.join(drivers_df, on= "driverId", how = "left", lsuffix ="_res", rsuffix = "_dri")
# check if it worked
resultsxdrivers_df.head(50)
```

```python
# group by drivers and sum descending
drivers_ranked = resultsxdrivers_df.groupby("driverId")["points"].sum().sort_values(ascending=False).reset_index()
```

```python
drivers_ranked.head(10)
```

```python
# merge with drivers_df to assign names
drivers_ranked_top10 = drivers_ranked.merge(drivers_df[['driverId', 'driverRef']], on='driverId', how='inner').head(10)
```

```python
drivers_ranked_top10.drop(columns=['driverId'])
```

```python
# create bar chart
drivers_ranked_top10.drop(columns=['driverId']).set_index('driverRef').plot.barh(figsize=(10,6))
```
