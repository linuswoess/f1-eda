Task 2: Find the top 10 drivers by total career points. Plot it as a bar chart.

```python
# join results with drivers
resultsxdrivers_df = results_df.join(drivers_df, on= "driverId", how = "left", lsuffix ="_res", rsuffix = "_dri")
# check if it worked
resultsxdrivers_df.head(50)
```

```python
# group by drivers and sum descending
drivers_ranked = resultsxdrivers_df.groupby("driverId")["points"].sum().sort_values(ascending=False).reset_index()
```

```python
drivers_ranked.head(10)
```

```python
# merge with drivers_df to assign names
drivers_ranked_top10 = drivers_ranked.merge(drivers_df[['driverId', 'driverRef']], on='driverId', how='inner').head(10)
```

```python
drivers_ranked_top10 = drivers_ranked_top10.drop(columns=['driverId'])
```

```python
# check drivers_ranked_top10
drivers_ranked_top10.head(10)
```

```python
# fix driver names in driverRef column
drivers_ranked_top10["driverRef"] = drivers_ranked_top10["driverRef"].replace(["hamilton", "vettel", "max_verstappen", "alonso", "raikkonen", "bottas", "rosberg", "michael_schumacher", "perez", "ricciardo"], ["L. Hamilton", "S. Vettel", "M. Verstappen", "F. Alonso", "K. Raikkonen", "V. Bottas", "N. Rosberg", "M. Schumacher", "S. Perez", "D. Ricciardo"])

# Bug Fix
# I originally wanted to change the names in column driverRef by using the following code: 
# drivers_ranked_top10 = drivers_ranked_top10["driverRef"].replace(["hamilton", "vettel", "max_verstappen", "alonso", "raikkonen", "bottas", "rosberg", "michael_schumacher", "perez", "ricciardo"], ["L. Hamilton", "S. Vettel", "M. Verstappen", "F. Alonso", "K. Raikkonen", "V. Bottas", "N. Rosberg", "M. Schumacher", "S. Perez", "D. Ricciardo"])
# which overwrote the entire dataframe with one column (i.e., a data series)
# to fix it, I had to assign the results back to only the desired column and not the whole data frame.
```
