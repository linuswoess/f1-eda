```python
# merge results with drivers
resultsxdrivers_df = results_df.merge(drivers_df, on= "driverId", how = "left", validate = "m:1", indicator = True, suffixes = ("_res", "_dri"))
# check if it worked
resultsxdrivers_df.head(50)
```

```python

# LESSON: merging dataframes
#
# STEP 1 — know your key relationship before merging
#
#   ask: "can this key appear multiple times in each df?"
#   run: print(df["key_col"].duplicated().sum())
#         0 = "1" side,  >0 = "m" side
#
#   "1:1" → one row per key on both sides
#   "m:1" → many rows on left, one row on right  (most common)
#   "1:m" → one row on left, many rows on right
#   "m:m" → many on both — usually a mistake, avoid
#
# STEP 2 — always use validate + indicator=True
#
#   merged_df = left_df.merge(
#       right_df,
#       on="key_col",
#       how="left",        # be intentional: inner/left/right/outer
#       validate="m:1",    # states your assumption, crashes if wrong
#       indicator=True,    # adds _merge column to check match quality
#   )
#
# STEP 3 — inspect the _merge column immediately after
#
#   print(merged_df["_merge"].value_counts())
#
#   "both"       → matched in both dfs        (expected)
#   "left_only"  → no match found in right df (investigate)
#   "right_only" → no match found in left df  (investigate)
#
#   filter unmatched rows:
#   merged_df[merged_df["_merge"] == "left_only"]
#
# STEP 4 — drop _merge when done
#
#   merged_df = merged_df.drop(columns="_merge")
#
# WHAT TO EXPECT FROM _merge BY JOIN TYPE
#
#   how="inner"  → "both" only
#   how="left"   → "both" + "left_only" expected
#                  "right_only" should NEVER appear — if it does, something is wrong
#   how="right"  → "both" + "right_only" expected
#                  "left_only" should NEVER appear — if it does, something is wrong
#   how="outer"  → all three are possible
```

**Day 2:** 
Which constructors have the most wins since 2010? 

```python
# load constructors.csv
constructors_df = pd.read_csv("data/constructors.csv")
```

```python
# inspect head of constructors_df
constructors_df.head()
```

```python
# inspect info of constructors_df
constructors_df.info()
```

```python
# merge constructors_df with results_df
constructxresults_df = constructors_df.merge(results_df,how = "left", on= "constructorId", suffixes = ("_con", "_res"))
```

```python
# inspect constructxresults_df
constructxresults_df.head()
```

```python
# drop columns not needed
constructxresults_df = constructxresults_df.drop(["url", "resultId", "driverId", "grid", "positionText", "statusId", "fastestLapSpeed", "fastestLapTime", "rank", "fastestLap", "milliseconds","time", "laps" ], axis = 1)
```

```python
# inspect constructxresults_df after dropping
constructxresults_df.head()
```

```python
# .drop() - two ways to drop columns:
#
# Option 1 - columns= (preferred, explicit):
# df.drop(columns=['driverId'])
#
# Option 2 - labels= with axis (generic):
# df.drop(labels='driverId', axis=1)  # axis=1 = columns
# df.drop(labels=0, axis=0)           # axis=0 = rows
#
# To drop a specific row by index:
# df.drop(labels=4, axis=0)  # drops row 4
# df.drop(4)                 # shorthand, same result
#
# axis=0 = rows (goes down)
# axis=1 = columns (goes across)
#
# Rule: always use columns= when dropping columns.
# Only use labels= when dropping rows by index.
```

```python
# now merge constructxresults_df with races_df
con_res_race_df = constructxresults_df.merge(races_df[["raceId", "year"]], on= "raceId", how = "inner")
```

```python
# Merging with only specific columns from another dataframe:
#
# df1.merge(df2[['key_col', 'wanted_col']], on='key_col', how='inner')
#
# Always include the key column (the one you're joining on)
# plus any other columns you want to bring in.
#
# Why double brackets [[]]?
# Single brackets df['col'] returns a Series (one column, no longer a dataframe)
# Double brackets df[['col']] returns a DataFrame (keeps table structure)
# You need a DataFrame going into merge, not a Series.
#
# Example:
# results_df.merge(drivers_df[['driverId', 'driverRef']], on='driverId', how='inner')
```

```python
# inspect con_res_race_df
con_res_race_df.head()
```

```python
# filter for only races since 2010 and position 1
con_res_race_2010_df = con_res_race_df[(con_res_race_df["year"] >=2010) & (con_res_race_df["position"] == "1")]
```

```python
# Filtering rows in a dataframe:
#
# WRONG - this returns True/False for each row, not the filtered dataframe:
# df["year"] >= 2010
#
# CORRECT - wrap it in outer brackets to actually filter:
# df[df["year"] >= 2010]
#
# The outer brackets say "give me all rows where this condition is True"
#
# Multiple conditions:
# df[(df["year"] >= 2010) & (df["position"] == 1)]  # AND
# df[(df["year"] >= 2010) | (df["position"] == 1)]  # OR
#
# Note: use & and | not 'and' and 'or' in pandas filters
# Each condition must be wrapped in its own brackets ()
```

```python
# inspect con_res_race_2010_df
con_res_race_2010_df.head()
```

```python
# group by constructors x points
best_construct_since2010_df = con_res_race_2010_df.groupby(["constructorId", "name"])["points"].sum().sort_values(ascending=False).reset_index()
```

```python
# Groupby multiple columns to preserve name columns:
#
# INEFFICIENT - groupby ID, then merge names back in:
# df.groupby("constructorId")["points"].sum()
# ...then merge constructors_df to get names again
#
# EFFICIENT - include the name column in groupby directly:
# df.groupby(["constructorId", "constructorRef"])["points"].sum()
#
# Both columns are preserved in the output, no extra merge needed.
#
# Rule: if the name column is already in your dataframe,
# include it in groupby instead of merging it back in afterwards.
```

```python
# reset_index() - what it does and when to use it:
#
# After groupby().sum(), the groupby column becomes the index (row label)
# instead of a regular column. This makes it hard to use afterwards.
#
# Without reset_index():
# constructorId         <- this is the index, not a column
#     1          4540.5
#     2          3098.0
#
# With reset_index():
#    constructorId  points   <- constructorId is now a regular column
# 0              1  4540.5
# 1              2  3098.0
#
# Rule: always add reset_index() after groupby() if you need to 
# merge on the grouped column or use it in further operations.^
```

```python
# inspect best_construct_since2010_df 
best_construct_since2010_df.head()
```

```python
# sort it again
best_construct_since2010_df = best_construct_since2010_df.sort_values("points", ascending = True)

# graph it
best_construct_since2010_df.plot.barh("name", ylabel = "Constructor", xlabel = "Total Points", title = "Best Constructors since 2010", legend= None, figsize = (10,6))
```
